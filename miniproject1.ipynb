```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Food_Production.csv")
df.head()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=["Eutrophying emissions per 100g protein (gPO₄eq per 100 grams protein)"],inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Food product'] = label_encoder.fit_transform(df['Food product'])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(columns=["Total_emissions"])
y = df["Total_emissions"]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

```python
xtrain
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
# Step 8: Evaluate the model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(ytest, ypred)
mse = mean_squared_error(ytest, ypred)
r2 = r2_score(ytest, ypred)

# Print results
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared Score:", r2)
```
